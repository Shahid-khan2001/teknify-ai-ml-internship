```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Load dataset
dataset = pd.read_csv("w3task1_dataset.csv")

# Display the first 5 rows to check
dataset.head()
```

```python
x = dataset[["YearsExperience"]]
y = dataset["Salary"]
```

```python
print(x.head())
print(y.head())
```

```python
plt.scatter(x,y)
```

```python
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.scatter(x,y)
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

# Display first 5 rows of training set to verify
print("x_train:")
print(x_train.head())

print("\ny_train:")
print(y_train.head())
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
print("Slope (coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
experience_df = pd.DataFrame({'YearsExperience': [5]})
predicted_salary = model.predict(experience_df)
print(f"Predicted salary for 5 years of experience: {predicted_salary[0]:.2f}")
```

```python
plt.figure(figsize=(8,5))

# Scatter plot of actual data
plt.scatter(x_train, y_train, color='blue', label='Training Data')

# Regression line (best fit)
plt.plot(x_train, model.predict(x_train), color='red', label='Best Fit Line')

plt.title("Salary vs Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Predict on test data
y_pred_test = model.predict(x_test)

# Calculate R score
r2 = r2_score(y_test, y_pred_test)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred_test)

print(f"R score on test data: {r2:.4f}")
print(f"Mean Squared Error on test data: {mse:.2f}")
```
